## Tensor flow linear model

```python
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt


def generate_dataset():
    # data is generated by y=2x + e where 'e' is sampled from normal distribution
    x_batch=np.linspace(-1,1,101)
    y_batch=2 * x_batch + np.random.randn(*x_batch.shape)*0.3
    return x_batch, y_batch
```

```python
def linear_regression():
    x=tf.placeholder(tf.float32,shape=(None,),name='x') # 2(a) - define place holders
    y=tf.placeholder(tf.float32,shape=(None,),name='y')
    
    with tf.variable_scope('mlreg') as scope:
        w = tf.Variable(np.random.normal(), name='W')
        #build graph
        y_pred = tf.mul(w,x)       
        loss=tf.reduce_mean(tf.square(y_pred-y))
        
    return x,y,y_pred,loss
```

```python
def run():
    #pass
    x_batch,y_batch=generate_dataset() # step 1: Get dataset
    
    x,y,y_pred,loss=linear_regression()  # Step 2: define your learning algorithm
    optimizer=tf.train.GradientDescentOptimizer(0.1).minimize(loss) # Step 3 define optimizer
    
    init = tf.global_variables_initializer() # 4a variables initialiser
    with tf.Session() as sess:
        sess.run(init)
        
        feed_dict={x:x_batch,y:y_batch}
        
        for _ in range(30):# is this 30 epochs? possibly..
            loss_val,_ = session.run([loss,optimizer],feed_dict)
            print('loss:',loss_val.mean())
            
        y_pred_batch=session.run(y_pred,{x:x_batch})
    plt.figure(1)
    plt.scatter(x_batch,y_batch)
    plt.plot(x_batch,y_pred_batch)
    #plt.savefig('plot.png')
    plt.show
```

```python
run()
```
