```python
def softmax(x):
    """
    Compute softmax values for each sets of scores in x.
    
    Rows are scores for each class. 
    Columns are predictions (samples).
    """
    scoreMatExp = np.exp(np.asarray(x))
    return scoreMatExp / scoreMatExp.sum(0)
#np.argmax(softmax(y_pred[25]))
```

```python

def pd():
    '''éèēīíìóòōúūùāẹịọụiouâḅḍ'''
    nasals={'m':'M','n':'N'}
    lateral={'l':'L'}
    rhotic={'r':'R'}
    glides={'w':'W', 'nw':'NW'}
    liquids={}
    liquids.update(glides)
    liquids.update(lateral)
    liquids.update(rhotic)
    approximant={'y':'Y','w':'W'}
    sonorants={}
    sonorants.update(nasals)
    sonorants.update(liquids)
    sonorants.update(approximant)
    fricatives={'f':'F','v':'V','s':'S','z':'Z'}
    affricates={'ch':'TS','j':'DZ','h':'H'}
    plosives={'p':'P','b':'B','ḅ':'BB','t':'T','d':'D','ḍ':'DD','k':'K','g':'G','kp':'KP','gb':'GB'}
    obstruents={}
    obstruents.update(fricatives)
    obstruents.update(affricates)
    obstruents.update(plosives)
    consonants={}
    consonants.update(sonorants)
    consonants.update(obstruents)
    vowela_open_central={'a':'A','à':'A0','ā':'A1','á':'A2','à':'A0','á':'A2'}
    vowelo_openmid_back={'ọ':'OH','ọ̀':'OH0','ọ̄':'OH1','ọ́':'OH2','ọ̀':'OH0','ọ́':'OH2'}
    vowelo_closemid_back={'o':'O','ò':'O0','ō':'O1','ó':'O2','ò':'O0','ó':'O2'}
    vowelu_close_back={'u':'U','ù':'U0','ū':'U1','ú':'U2','ú':'U2'}
    vowelu_nearclose_nearback={'ụ':'UH','ù':'UH0','ụ̄':'UH1','ụ́':'UH2'}
    vowele_openmid_nearfront={'e':'E','è':'E0','ē':'E1','é':'E2'}
    vowele_closemid_nearfront={'ẹ':'EH','è̩':'EH0','ẹ̄':'EH1','ẹ́':'EH2'}
    voweli_close_front={'i':'I','ì':'I0','ī':'I1','í':'I2'}
    voweli_nearclose_nearfront={'ị':'IH','ị̀':'IH0','ị̄':'IH1','ị́':'IH2','ị́':'IH2','ị':'IH','ị̀':'IH0'}
    vowels={}
    vowels.update(vowela_open_central)
    vowels.update(vowelo_openmid_back)
    vowels.update(vowelo_closemid_back)
    vowels.update(vowelu_close_back)
    vowels.update(vowelu_nearclose_nearback)
    vowels.update(voweli_close_front)
    vowels.update(voweli_nearclose_nearfront)
    vowels.update(vowele_openmid_nearfront)
    vowels.update(vowele_closemid_nearfront)
    pd={}
    pd.update(consonants)
    pd.update(vowels)
    return pd  
def getPhones(word):
    toskip=False
    toret=[]
    for i in range(len(word)):
        try:
            if toskip:
                toskip=False
                continue
            if i+1<len(word) and word[i]+word[i+1] in pd():
                toskip=True
                toret.append(pd()[word[i]+word[i+1]])
            else:
                toret.append(pd()[word[i]])
        except:
            print(word[i],'=',word)
    return " ".join(toret)   
getPhones('love')
```
