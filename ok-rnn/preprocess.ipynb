```python
sx=[len(x) for x in iX]
sy=[len(x) for x in tY]

def seqpad(seq,seql):
    arr=[]
    m=max(seql)
    for i in seq:
        n=m-len(i)
        for j in range(n):
            i.append(-1)
        arr.append(i)
    return arr

def seqpad2(seq,m,ix):
    arr=[]
    for i in seq:
        n=m-len(i)
        for j in range(n):
            i.append(ix)
        arr.append(i)
    return arr

def seqpad3(iX,tY,cx2ix,cy2ix):
    nix=[]
    nty=[]
    bX=[]
    bY=[]
    buf=0
    c=0
    for i in range(len(iX)):
        #if(buf>40):
        #    print(c,':',buf)
        if len(iX[i])==len(tY[i]) and buf+len(iX[i])<50:
            iX[i].append(cx2ix['*'])
            tY[i].append(cy2ix['*'])
            nix.extend(iX[i])
            nty.extend(tY[i])
        elif len(iX[i])>len(tY[i]) and len(iX[i])+buf<50:
                #print((iX[i].append(cx2ix['*'])))
                [tY[i].append(cy2ix['*']) for j in range(abs(len(iX[i])-len(tY[i])))]
                iX[i].append(cx2ix['*'])
                tY[i].append(cy2ix['*'])
                nix.extend(iX[i])
                nty.extend(tY[i])
        elif len(iX[i])<len(tY[i]) and len(tY[i])+buf<50:
                [iX[i].append(cx2ix['*']) for j in range(abs(len(iX[i])-len(tY[i])))]
                iX[i].append(cx2ix['*'])
                tY[i].append(cy2ix['*'])
                nix.extend(iX[i])
                nty.extend(tY[i])
        else: #if buf=>50:
            buf=50-buf
            [nix.append(cx2ix['*']) for j in range(buf)]
            [nty.append(cx2ix['*']) for j in range(buf)]
            bX.append(nix)
            bY.append(nty)
            nix=[]
            nty=[]
            buf=0
            c+=1
        buf=len(nix)
    return (bX,bY)
pX,pY=seqpad3(iX,tY,cx2ix,cy2ix)
pX=np.array(pX)
pY=np.array(pY)
print(pY.shape,pX.shape)
```

```python
def seq2sent(seq, dox=False):
    return "".join([(ix2cx[i] if dox else ix2cy[i]) for i in seq]).replace('*',' ')
```

```python
np.random.seed(42)
def split_train_test(x,y, test_ratio):
    shuffled_indices=np.random.permutation(len(x))
    test_set_size=int(len(x)*test_ratio)
    test_indices=shuffled_indices[:test_set_size]
    train_indices=shuffled_indices[test_set_size:]
    return x.iloc[train_indices],x.iloc[test_indices],y.iloc[train_indices],y.iloc[test_indices]

xtr, xts,ytr,yts = split_train_test(pd.DataFrame(pX),pd.DataFrame(pY), 0.2)
iX = xtr.values
tY = ytr.values
print(len(tY),len(yts.values))
```

```python
import tensorflow as tf
reset_graph()

n_steps = 50
n_neurons = 200
n_layers = 3
num_encoder_symbols = len(cx) #20000
num_decoder_symbols = len(cy) #20000
embedding_size = 1
learning_rate = 0.001

X = tf.placeholder(tf.int32, [None, n_steps]) # English sentences
Y = tf.placeholder(tf.int32, [None, n_steps]) # French translations
#y = tf.placeholder(tf.int32,[None])
w = np.ones( (len(iX), n_steps - 1, 1), dtype=np.float32 )
W = tf.Variable(w,name='W')

Y_input = Y[:, :-1]
Y_target = Y[:, 1:]
W_flat = tf.reshape(W, [-1])

encoder_inputs = tf.unstack(tf.transpose(X)) # list of 1D tensors
decoder_inputs = tf.unstack(tf.transpose(Y_input)) # list of 1D tensors

lstm_cells = [tf.contrib.rnn.BasicLSTMCell(num_units=n_neurons)
              for layer in range(n_layers)]
cell = tf.contrib.rnn.MultiRNNCell(lstm_cells)

output_seqs, states = tf.contrib.legacy_seq2seq.embedding_rnn_seq2seq(
#output_seqs, states = tf.contrib.legacy_seq2seq.basic_rnn_seq2seq(
    # X, 
    encoder_inputs,
    # Y_input, 
    decoder_inputs,
    cell,
    num_encoder_symbols,num_decoder_symbols,embedding_size
    ,feed_previous=True 
    #,output_projection=(W_flat,_)
    )
# encoder_outputs: [max_time, batch_size, num_units]
logits = tf.transpose(tf.unstack(output_seqs), perm=[1, 0, 2])

logits_flat = tf.reshape(logits, [-1, num_decoder_symbols])

#logits_flat = tf.layers.dense(states,num_decoder_symbols)
Y_target_flat = tf.reshape(Y_target, [-1])
xentropy = W_flat * tf.nn.sparse_softmax_cross_entropy_with_logits(labels=Y_target_flat, logits=logits_flat)
loss = tf.reduce_mean(xentropy)
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
training_op = optimizer.minimize(loss)

correct = tf.nn.in_top_k(logits_flat, Y_target_flat, 1)
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
n_epochs = 501
#batch_size = 150
tne=[]
tte=[]
with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        #for iteration in range(mnist.train.num_examples // batch_size):
        #    X_batch, y_batch = mnist.train.next_batch(batch_size)
        #    X_batch = X_batch.reshape((-1, n_steps, n_inputs))
        sess.run(training_op, feed_dict={X: iX, Y: tY})
        acc_train = accuracy.eval(feed_dict={X: iX, Y: tY})
        acc_test = accuracy.eval(feed_dict={X: xts.values, Y:yts.values})
        if epoch%10==0 or epoch==n_epochs-1:
            print("Epoch", epoch, "Train accuracy:", acc_train, "Test accuracy:", acc_test)
            tne.append(1-acc_train)
            tte.append(1-acc_test)
        if acc_test>0.9:
            break
    saver.save(sess,"./okd_model")
```

```python
def predict(x):
    x=np.ravel(seqpad2([[cx2ix[ch] for ch in x]],50,cx2ix['*']))
    X_new = np.array([x])
    ypad = np.array([np.ravel(seqpad2([[cy2ix['*']]],50,cy2ix['*']))])
    with tf.Session() as sess:                          # not shown in the book
        saver.restore(sess, "./okd_model")   # not shown
        y_pred = sess.run(logits_flat, feed_dict={X: X_new, Y:ypad })
        x=tf.nn.top_k(y_pred,1)
        t=sess.run(x)
    return seq2sent([i[0] for i in t.indices])
```
